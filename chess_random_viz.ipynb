```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.ticker as ticker
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme()
```

```python
# move counts
with open ('random_move_counts.pickle', 'rb') as f:
    movecounts = pickle.load(f)
movecounts = pd.DataFrame.from_dict(movecounts, orient='index').reset_index()
movecounts.columns = ['Move Count', 'Occurrences']
movecounts.sort_values(by=['Move Count'])
```

```python
f, ax = plt.subplots(figsize=(15, 6))
sns.set_color_codes("muted")
sns.barplot(x='Move Count', y='Occurrences', data=movecounts, color="b", edgecolor="none")
ax.xaxis.set_major_locator(ticker.MultipleLocator(25))
ax.xaxis.set_major_formatter(ticker.ScalarFormatter())
plt.tight_layout()
fig = ax.get_figure()
fig.savefig('random_move_counts.png', dpi=800)
```

```python
with open ('random_chess_results.pickle', 'rb') as f:
    random_results = pickle.load(f)
random_results = pd.DataFrame.from_dict(random_results, orient='index').reset_index()
random_results.columns = ['Conclusion', 'Occurrences']
random_results = random_results.sort_values(by=['Occurrences'], ascending=False, axis=0)

f, ax = plt.subplots(figsize=(7, 9))
sns.set_color_codes("muted")
sns.barplot(y='Conclusion', x='Occurrences', data=random_results, color="g", edgecolor="none", orient='h')
plt.tight_layout()
fig = ax.get_figure()
fig.savefig('random_move_results.png', dpi=800)
```
